4. ***Create User-Defined Function for Price Ranges***

```python
def price_label(row):
    if row['prices']<=5:
        return 'Low-Range Product'
    elif (row['prices']>5) and (row['prices']<=15):
        return 'Mid-Range Product'
    elif row['prices']>15:
        return 'High-Range Product'
    else:return 'Not enough Data'
```

```python
#Applying to the dataframe
df['price_range']=df.apply(price_label, axis=1)
```

Observation:  This is not an error message.  It's a heads up from Python that there is a more efficient way to create a function for price ranges and suggests to use the loc() function instead.

```python
#Use For-Loop to filter order days by Busiest, Least and Regularly Busy
result=[]
for value in ords_prods_merge['order_day_of_week_name']:
    if value =='Sunday':
        result.append('Busiest Day')
    elif value=='Thursday':
        result.append('Least Busy')
    else:
        result.append('Regularly Busy')
```

7.2 ***Create new column "Busiest_Days" :  Identify TWO busiest days of week and TWO slowest days of week***

```python
#Printing frequency of orders by day of the week
ords_prods_merge['order_day_of_week_name'].value_counts(dropna=False)
```

```python
#Will Use For-Loop to filter order days of the week by the TWO busiest and TWO least busiest days of week.
#Needed to add "in" operator as well as use brackets because there are TWO for each: Busiest and Least Busiest Days of the week.
result = []
for value in ords_prods_merge['order_day_of_week_name']:
    if value in ['Sunday', 'Monday']:
        result.append('Busiest Days')
    elif value in ['Wednesday', 'Thursday']:
        result.append('Slowest Days')
    else:
        result.append('Regularly Busy Days')
```

```python
result
```

```python
#Creating new column in ords_prods_merge dataframe: "Busiest Days"
ords_prods_merge['Busiest_Days']=result
```
