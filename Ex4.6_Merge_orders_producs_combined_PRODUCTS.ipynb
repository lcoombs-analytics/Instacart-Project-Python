5. ***Merging DataFrames***

```python
#Testing merge without overwriting DataFrames using merge flag 'match'
pd.merge(df_orders_products_combined , df_prods , on = ['product_id'] , how = 'inner', indicator = 'match' )
```

Observation: Both DataFrames are a full match with the 'product-id' column. 

```python
#Merging df_merged_large_prods and df_prods_prior DataFrames
df_merged_large_prods = df_orders_products_combined.merge(df_prods, on='product_id')
```

```python
df_merged_large_prods['_merge'].value_counts()
```

Observation:  Inner join is appropriate for these the merge
